```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import datetime
```

```python
def list_dates(start,end):
    """ This creates a list of of dates between the 'start' date and the 'end' date """
    # create datetime object for the start and end dates
    start = datetime.datetime.strptime(start, '%Y-%m-%d')
    end = datetime.datetime.strptime(end, '%Y-%m-%d')
    # generates list of dates between start and end dates
    step = datetime.timedelta(days=1)
    dates = []
    while start <= end:
        dates.append(start.date())
        start += step
    # return the list of dates in string format
    return [str(date) for date in dates]
```

```python
#change dates accordingly
start = '2014-1-1'
end = '2014-12-31'
dates = list_dates(start,end)
```

```python
driver = webdriver.Chrome('C:/Windows/System32/chromedriver')
```

```python
from collections import OrderedDict
```

```python
#frame=[]
frame=OrderedDict() 
```

```python
for date in dates:
    driver.get('https://www.wunderground.com/history/daily/in/aluva/VOCI/date/' +date)
    tables = WebDriverWait(driver,5).until(EC.presence_of_all_elements_located((By.XPATH,'//*[@id="history-observation-table"]')))
    for table in tables:  ###tables contain the table which is present in xpath
        newTable = (pd.read_html(table.get_attribute('outerHTML')))
        #frame.append(newTable[0])
        newTable[0].drop(['Wind','Precip.','Wind Gust','Precip Accum','Condition','Wind Speed'],axis=1,inplace=True)
        frame[date]=newTable[0]
        frame[date]['Date']=date
```

```python
weather_data=[]
from datetime import *
for date in dates:
    frame[date]['Time']=[datetime.strftime(datetime.strptime(t, '%I:%M %p'), "%H:%M") for t in frame[date]['Time']]
    observation=frame[date].sort_values(by=['Time'])
    weather_data.append(observation)
weather_data=pd.concat(weather_data,ignore_index=True)
    
```
